```python
# importing the necessary libraries
!pip install matplotlib
!pip install nltk
!pip install wordcloud
!pip install scipy
!pip install scikit-learn

import matplotlib.pyplot as plt
import nltk
import string
import warnings
from scipy.stats import pearsonr
from nltk.corpus import stopwords
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Download NLTK stopwords
nltk.download("stopwords")

# Ignore warnings
warnings.filterwarnings('ignore')
```

```python
df.isnull().sum()
```

```python
splitted = df['posted'].str.split(' ', expand=True)

# create new columns
df['year'] = splitted[2].astype('int')
df['month'] = splitted[1]
```

```python
df['year'].value_counts().plot.bar()
plt.show()
```

```python
# remove irrelevant information

df['details'] = df['title'] + ' ' + df['details']

df = df[['main_speaker', 'details']]
df.dropna(inplace=True)
df.head()
```

```python
# make a copy of the data

data = df.copy()
```

```python
# text processing

def remove_stopwords(text):
    stop_words = stopwords.words('english')
    
    imp_words = []
    for word in str(text).split():
        word = word.lower()
        if word not in stop_words:
            imp_words.append(word)
    output = " ".join(imp_words)
    return output

```

```python
df['details'] = df['details'].apply(lambda text: remove_stopwords(text))

df.head()
```

```python
# remove punctuations

punctuations_list = string.punctuation

def cleaning_punc(text):
    signal = str.maketrans('', '', punctuations_list)
    return text.translate(signal)

df['details'] = df['details'].apply(lambda x: cleaning_punc(x))
df.head()
```

```python
# use wordcloud to visualize word occurence

details_corpus = " ".join(df['details'])

plt.figure(figsize=(20,20))
wc = WordCloud(max_words=1000,
               width=800,
               height=400).generate(details_corpus)

plt.axis('off')
plt.imshow(wc)
plt.show()
```

```python
# building the unsupervised machine learning recommendation system
vectorizer = TfidfVectorizer(analyzer = 'word')
vectorizer.fit(df['details'])
```

```python
df.head()
```

```python
# cosine similarity and pearson correalation

def get_similarities(talk_content, data=df):
    talk_array = vectorizer.transform(talk_content).toarray()
    sim = []
    pea = []
    for idx, row in data.iterrows():
        details = row['details']
        # get the vector for current talks
        talk_arrays = vectorizer.transform(
            data[data['details'] == details]['details']).toarray()
        # using as cosine sim
        cos_sim = cosine_similarity(talk_array, talk_arrays)[0][0]
        # pearson corr
        pea_sim = pearsonr(talk_array.squeeze(), talk_arrays.squeeze())[0]
        sim.append(cos_sim)
        pea.append(pea_sim)
    return sim,pea
```

```python
def recommend_talks(talk_content, data=data):
    data['cos_sim'], data['pea_sim'] = get_similarities(talk_content)
    data.sort_values(by=['cos_sim', 'pea_sim'], ascending=[False, False], inplace=True)
    display(data[['main_speaker', 'details']].head())
```

```python
talk_content = ['Time Management and working\
hard to become successful in life']

recommend_talks(talk_content)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("tedx_datase.csv")   # change filename if needed

# Show first few rows and column names
print(df.head())
print(df.columns)

# Drop rows where num_views is missing
df = df.dropna(subset=["num_views"])

# Select numeric feature(s) for clustering
X = df[["num_views"]]

# Run KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df["cluster"] = kmeans.fit_predict(X)

# Show cluster distribution
print("\nCluster counts:")
print(df["cluster"].value_counts())

# Plot bar ch
```

```python
# Create new numeric feature: length of 'details'
df["details_length"] = df["details"].apply(lambda x: len(str(x)))

# Select numeric feature(s) for clustering
X = df[["details_length"]]

# Run KMeans clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
df["cluster"] = kmeans.fit_predict(X)

# Show cluster distribution
print("Cluster counts:")
print(df["cluster"].value_counts())

# Plot clusters
import matplotlib.pyplot as plt

df["cluster"].value_counts().plot(kind="bar")
plt.title("Cluster Distribution (based on details length)")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.show()

print("Hello from clustering")
```

```python
# Look at average details_length per cluster
print("\nAverage details length by cluster:")
print(df.groupby("cluster")["details_length"].mean())

# Look at sample talks from each cluster
for c in df["cluster"].unique():
    print(f"\nCluster {c} sample talks:")
    print(df[df["cluster"] == c][["main_speaker", "title", "details_length"]].head(5))
```
